## 4. Visualise Data

### 4.1 Visualise Data for Nigeria

#### Extract data for Nigeria

```python
# Let Nigeria = NG
NG = CC.query("Country =='Nigeria'")
NG
```

#### Plot the 'NG' Data on Graph

```python
country_of_interest = NG.loc[103] # NG index number = 106

country_name = country_of_interest['Country']
years = country_of_interest.index[1:] # Exclude the 'Country' column
temp = country_of_interest.values[1:] # Exclude the 'Country' column values


plt.figure(figsize = (10,6))
plt.plot(years, temp)

plt.title(f'Average Temperature in {country_name} (1961-2022)')
plt.xlabel('Year')
plt.ylabel('Average Temperature (áµ’C)')
plt.xticks(rotation = 90) # Rotate x-axis ticks by 90 degrees
plt.legend()
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent overlap

# Save figure
plt.savefig('Average_temperature_in_Nigeria(1961-2022).png')
plt.show()
```

### 4.2 Compare Nigeria vs Some Other Countries

```python
# List of countries to compare
countries_compared = ['Austria', 'Bangladesh', 'Nigeria']

# Filter the dataframe to include only selected countries
CC_compared = CC[CC['Country'].isin(countries_compared)]

# Plot
plt.figure(figsize=(10,6))

for index, row in CC_compared.iterrows():
    country = row['Country']
    years = row.index[1:] # Exclude the 'Country' column
    temp = row.values[1:] # Exclude the 'Country' column values
    plt.plot(years, temp, label= country)

plt.title('Compparison of Average Temperature by Countries (1961-2022)')
plt.xlabel('Year')
plt.ylabel('Average Temperature (áµ’C)')
plt.xticks(rotation = 90) # Rotate x-axis ticks by 90 degrees
plt.legend()
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent overlap

# Save figure
plt.savefig('temperature_comparision_Austria_Bangladesh_Nigeria(1961-2022).png')
plt.show()
```

### 4.3 Checking for Country with the Highest Temperature in a Particular Year

```python
# Select the column for the year to check
year_to_check = 'F2022' # Can be changed to any other desired year

# Get the index of the country with the highest temperature for the year
country_with_highest_temp = CC.set_index('Country')[year_to_check].idxmax()

# Get the max temp value for the year
highest_temp = CC.set_index('Country')[year_to_check].max()

print(f'The country with the highest temperature in {year_to_check} is {country_with_highest_temp} with a temperature of {highest_temp}C')
```

#### Top 10 Countries with Highest Temperatures Visualisation

```python
# Select the column for the year to check
year_to_check = 'F2022' # Can be changed to any other desired year

# Sort the DataFrame by the temp for the selected year in descending order
sorted_CC = CC.sort_values(by = year_to_check, ascending = False)

# Select the top 10 countries with the highest temperatures
top_10_countries = sorted_CC.head(10)

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(top_10_countries['Country'], top_10_countries[year_to_check], color = 'skyblue')

# Annotate the values on the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 3), ha = 'center', va = 'bottom')

plt.title(f'Top 10 Countries with the Highest Temperature in {year_to_check}')
plt.xlabel('Country')
plt.ylabel('Average Temperature (áµ’C)')
plt.xticks(rotation = 45, ha = 'right') # Rotate x-axis ticks by 90 degrees
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent overlap

# Save figure
plt.savefig('top 10 countries with highest temp in F2022.png')
plt.show()
```

### 4.5 Checking the Behaviour of temperatures in the Top 10 Countries in a Period of 5 Years (2018-2022)

```python
# Select the years you want to check
years_to_check = ['F2022', 'F2021', 'F2020', 'F2019', 'F2018']  # Change to the desired years

# Sort the DataFrame by the specified years in descending order
sorted_CC = CC.sort_values(by=years_to_check, ascending=False)

# Select the top 10 countries with the highest temperatures
top_10_countries = sorted_CC.head(10)

# Plot
plt.figure(figsize=(12, 8))

# Plot bars for each year for each country
for i, year in enumerate(years_to_check):
    bars = plt.bar([x + i * 0.1 for x in range(len(top_10_countries))],
                   top_10_countries[year],
                   width=0.1,
                   label=year)

    # Annotate the values on the bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2), ha='center', va='bottom')

plt.title(f'Top 10 Countries with the Highest Temperatures ({", ".join(years_to_check)})')
plt.xlabel('Country')
plt.ylabel('Average Temperature (Â°C)')
plt.xticks([r + 0.3 for r in range(len(top_10_countries))], top_10_countries['Country'], rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()

# Save figure
plt.savefig('comparison of top 10 relative to 5 years.png')
plt.show()
```
